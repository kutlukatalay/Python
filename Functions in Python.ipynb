### NESTED FUNCTIONS

def outer():
    """..."""
    x = ...
    
    def inner():
    """..."""
    y = x ** 2
    
    return ...

### X referansı inner'dan alıyor. Python önce local scope inner'da arıyor, sonra bulamazsa outer'da. Bulamazsa global ve built-in scopelara bakıyor.

```python
def mod2plus5(x1,x2,x3):
    
    def inner(x):
        
        return x % 2 + 5
    
    return (inner(x1),inner(x2),inner(x3))
```

```python
print(mod2plus5(1,2,3))
```

### Burada önce inner fonksiyon çalıştı.  1,2,3 değerleri inner fonksiyonun çalışması sonucunda 6,5,6 değerlerini verdi.

---

```python
def raise_val(n):
    
    def inner(x):
        
        raised = x ** n
        return raised
    
    return inner
```

```python
square = raise_val(2)

cube = raise_val(3)

print(square(2),cube(4))
```

### Burada raise_val fonksiyonu n argümanını alıyor. Bu değeri dışardan biz veriyoruz. inner adında başka bir fonksiyon tanımladık. Bu fonksiyon, dışarıdaki raise_val fonksiyonunun parametresi olan n'yi kullanabilir. Yani, inner fonksiyonu raise_val fonksiyonunun kapsamında (scope) bulunan bir değişkene (n) erişebilir.

### inner fonksiyonu, kendisine parametre olarak verilen x sayısının n'inci kuvvetini hesaplar (x ** n) ve bu değeri raised adlı bir değişkende saklar. Son olarak, inner fonksiyonu, hesaplanan raised değerini döndürür. raise_val fonksiyonu ise inner fonksiyonunu döndürür. Yani, raise_val fonksiyonu çağrıldığında, aslında dışarıya inner fonksiyonu geri döner.

---

```python
def outer():
    """Prints the value of n."""
    n = 1
    
    def inner():
        nonlocal n
        n = 2
        print(n)
        
    inner()
    print()
```

```python
outer()
```

### nonlocal parametresi dışardaki fonksiyonda tanımlı değeri değiştirir.

```python
def echo_shout(word):
    """Change the value of a nonlocal variable"""
    
    
    echo_word = word + word

    print(echo_word)
    
    
    def shout():
        """Alter a variable in the enclosing scope"""    
        
        nonlocal echo_word
        
        echo_word = echo_word + '!!!'
    
    
    shout()
    
    
    print(echo_word)


echo_shout('hello')
```

## ERROR HANDLING

```python
def sqrt(x):
    """Returns the square root of x"""
    try:
        return x ** 0.5
    except:
        print("x must be an int or float")
```

### Burada try ve except eklememizin asıl mantığı fonksiyonun istediğimiz şekilde çalışması. Try istediğimiz işlem için verilen kısıt, except ise hata aldığımızda fonksiyonun vereceği mesaj

## LIST COMPREHENSIONS

## Elimizde sayılardan oluşan bir liste olduğunu varsayalım. Bu listeyi kullanarak yeni bir liste oluşturmak istiyoruz.

```python
nums = [12,8,21,3,16]

new_nums = []

for i in nums:
    new_nums.append(i + 1)
    
print(new_nums)
```

### Bu işlemi for loopla yapmak yerine list comprehension ile yapabiliriz.

```python
nums = [12,8,21,3,16]

new_nums2 = [i + 1 for i in nums]

print(new_nums2)
```
